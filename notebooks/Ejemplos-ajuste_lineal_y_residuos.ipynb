```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Ejemplo 1: modelo lineal + error simulado

```python
N = 70
x = np.linspace(0,10,N)
y = x + 0.2*(np.random.rand(N)-0.5)
```

```python
b1, b0 = np.polyfit(x,y,1)
xx = np.linspace(-1,11,10000)
plt.scatter(x,y)
plt.plot(xx,b0+b1*xx, color='red', linewidth=2)
```

```python
# Coeficiente de Determinación
St = np.sum((y-np.mean(y))**2)
Sr = np.sum((y-b0-b1*x)**2)
r2 = (St-Sr)/St
print(r2)
```

```python
# coeficiente de posición y pendiente:
print(b0,b1)
```

```python
# Residuos
ep = y - (b0+b1*x)
# gráfico de residuos
plt.scatter(x,ep)
plt.grid()
```
